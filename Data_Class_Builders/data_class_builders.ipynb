### Использование dataclass

```python
from dataclasses import dataclass, field
from typing import ClassVar

@dataclass(frozen=True)
class ClubMember1:
    name: str
    members: ClassVar[list[str]] = list() # Общий список

    def __post_init__(self):
        if self.name is not None:
            self.members.append(self.name)

@dataclass
class ClubMember2:
    name: str
    members: list = field(default_factory=list) # У каждого свой список
```

```python
cm1 = ClubMember1("Q")
cm2 = ClubMember1("W")

print(cm2.members)
```

```python
cm1 = ClubMember2("Q")
cm2 = ClubMember2("W")

cm1.members.append(cm1.name)

print(cm2.members)
```
